```python
Q. Analyze Affordable/Budgeted and Highest Rated Restaurants of Bangalore:
```

```python
df_Affordable_Restaurants = swiggy_df[(swiggy_df['Cost_for_Two (in Rupees)'] <= 500) & (swiggy_df['Rating'] >= 4.0)]
df_Affordable_Restaurants
```

```python
df_Affordable_Restaurants = df_Affordable_Restaurants.groupby(['Shop_Name', 'Rating'])['Cost_for_Two (in Rupees)'].agg('mean')
df_Affordable_Restaurants = df_Affordable_Restaurants.reset_index()
df_Affordable_Restaurants
```

```python
df_Affordable_Restaurants.sort_values(by = ['Rating'], ascending = False, inplace = True)
df_Affordable_Restaurants
```

```python
plt.figure(figsize = (18, 7))
sns.barplot(x = df_Affordable_Restaurants['Shop_Name'], y = df_Affordable_Restaurants['Cost_for_Two (in Rupees)'],
           data = df_Affordable_Restaurants)
plt.title('Affordable/Budgeted and Highest Rated Restaurants ', fontsize = 14, fontweight = 'bold', 
          fontstyle = 'italic')
plt.xlabel('Shop_Name', fontsize = 10, fontweight = 'bold')
plt.xlabel('Approx. Cost_for_Two (₹)', fontsize = 10, fontweight = 'bold')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Cuisine Analysis
```

```python
swiggy_df['Cuisine'] = swiggy_df['Cuisine']
swiggy_df['Cuisine']
```

```python
swiggy_df
```

```python
swiggy_df['Cuisine'].unique()
```

```python
freq_dict={}
for i in swiggy_df['Cuisine'].unique():
    Cuisines_Lists = i.split(',')
    for Cuisine in Cuisines_Lists:
        Cuisine = Cuisine.lstrip(' ')
        if Cuisine in freq_dict:
            freq_dict[Cuisine] = freq_dict[Cuisine] + 1
        else: 
            freq_dict[Cuisine] = 1
        
print(freq_dict)
print()
print('Total Records: \t', len(freq_dict))
```

```python
Cuisine = freq_dict.keys()
freq = freq_dict.values()
```

```python
Cuisine_Analysis_df = pd.DataFrame()
```

```python
Cuisine_Analysis_df['Cuisine'] = Cuisine
Cuisine_Analysis_df['Count'] = freq
```

```python
Cuisine_Analysis_df
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=Cuisine_Analysis_df,x='Cuisine',y='Count')
plt.xticks(rotation=90)
plt.title('Cuisines Overall Analysis ', fontsize = 14, fontweight = 'bold', fontstyle = 'italic')
plt.show()
```
